```python
# Library
import pandas as pd
import random as rand
```

```python
# 1. select path
from google.colab import drive
drive.mount('/content/drive')
ruta_data = '/content/drive/MyDrive/Colab Notebooks/Data/grupo1626.xlsx'
ruta_data_2 = '/content/drive/MyDrive/Colab Notebooks/Data/PC - 1626.xlsx'
```

```python
# 2. Import data
grupo = pd.read_excel(ruta_data)
carreras = pd.read_excel(ruta_data_2, sheet_name="Hoja2")
grupo.head()
```

```python
carreras.head()
```

```python
# 3. Explore stundent's names with group

nombres = grupo["First Name"]
nombres
```

```python
# 4. Select indexes

grupo.set_index('Username', inplace=True)
grupo.head()
```

```python
carreras.set_index('Username', inplace=True)
carreras.head()
```

```python
# 4. concatenate two dataframes
final_data = pd.concat([grupo, carreras], axis=1)
final_data
```

```python
# 5. varify shape
final_data.shape
```

```python
#6. reset index
final_data = final_data.reset_index()
final_data
```

```python
# 7 add final grade

grade_rand = {'final_grade':[rand.randint(0,5) for i in range(len(grupo))]}

#convertir el diccionario a df
df_grade = pd.DataFrame(grade_rand)
df_grade

```

```python
# 8. concatenar
final_data = pd.concat([final_data,df_grade], axis=1)
final_data
```

```python
# 9. Nota promedio por curso

promedio_curso = final_data.groupby('carrera')['final_grade'].mean()
promedio_curso_sorted = promedio_curso.sort_values(ascending=False)
print(promedio_curso_sorted)
```
